```python
import stanza
```

```python
#downlaod the model
stanza.download('en')
```

```python
#tokenization
staza_token =stanza.Pipeline('en' , processors='tokenize')
```

```python
a = staza_token('Barack Obama was born in hawaii . I was born in india')
a
```

```python
for i , sentence in enumerate(a.sentences):
    print(f"============Setence {i + 1} tokens ==================")
    print(*[f'id:{ token.id}\ttext: {token.text}' for token in sentence.tokens] , sep = '\n')
```

```python
staza_token =stanza.Pipeline('en' , processors='tokenize' , use_gpu=True)
```

```python
for i in a.sentences:
    print(i)
    for j in i.words:
        print(j.text)
```

```python
staza_token =stanza.Pipeline('en', use_gpu=True)
```

```python
a = staza_token('Barack Obama was born in hawaii . I was born in india')
a
```

```python
for i , sentence in enumerate(a.sentences):
    print(f'==============Sentence{i+1} tokens ================')
    print(*[f'id: {token.id}\ttext: {token.text}\tupos:{token.upos}\tlemma:{token.lemma}'for token in sentence.words] , sep='\n')
```

```python
from spacy_stanza import StanzaLanguage
from spacy import displacy
```

```python
abc = stanza.Pipeline('en' , use_gpu=True)
```

```python
nlp = StanzaLanguage(abc)
doc = nlp('Barack obama was born in hawaii, I was born in india')
```

```python
displacy.render(doc , style='dep')
```

```python
displacy.render(doc , style='ent')
```
